```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Product': ['A', 'B', 'A', 'C', 'B', 'A'],
    'Region': ['East', 'West', 'East', 'North', 'West', 'South'],
    'Sales': [100, 150, 120, 90, 180, 110]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
display(df)

# Filter data for 'Product A'
product_a_sales = df[df['Product'] == 'A']

print("\nSales for Product A:")
display(product_a_sales)

# Calculate total sales per region
regional_sales = df.groupby('Region')['Sales'].sum().reset_index()

print("\nTotal Sales per Region:")
display(regional_sales)
```
